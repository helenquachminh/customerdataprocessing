```python
df = df.dropna(how='any', axis=0)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
import matplotlib.pyplot as plt
```

```python
#Let's visualise the number of application in our session (per session)
```

```python
application_count = df[df['event_action']=='application']
application_count
```

```python
grouped_application_count = application_count.groupby('session_id')['event_action'].count()
grouped_application_count
```

```python
session_application_counts = grouped_application_count.value_counts().to_frame()
session_application_counts
```

```python
session_application_counts.plot(kind='bar',stacked=False)
plt.title('Number of sessions with an application')
plt.xlabel('Number of applications')
plt.ylabel('Number of sessions')

plt.show()
```

```python
session_application_counts.plot(kind='bar',stacked=False, color='green')
plt.title('Number of sessions with an application')
plt.xlabel('Number of applications')
plt.ylabel('Number of sessions')
plt.figure(figsize=(20,16))

plt.tight_layout()
plt.show()
plt.savefig('my_plot.png')
```

```python
new_df = pd.read_csv('/content/gdrive/MyDrive/Analysis_Python/Workshop_2.csv')
```

```python
new_df
```

```python
application_count = new_df[new_df['event_action']=='application']
application_count
```

```python
applications_per_session = application_count.groupby('session_id')['event_action'].count()
applications_per_session
```

```python
pageviews_per_session = new_df.groupby('session_id')['pageviews'].sum()
pageviews_per_session
```

```python
metrics = pd.DataFrame({'Application':applications_per_session, 'Pageviews': pageviews_per_session})
```

More visuals

```python
import seaborn as sns
```

```python
sns.scatterplot(data=metrics, x='Application', y='Pageviews', color = 'green')

plt.title('Relationship between Applications and Pageviews per Session')
plt.xlabel('Applications per session')
plt.ylabel('Pageviews per session')

plt.show()
```
